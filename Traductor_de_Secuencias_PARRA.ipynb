**Pasos para generar la aplicacion**

1. Construir una funcion con el diccionario codon --> aminoacido.

2. Construir la funcion para leer archivo.
3. Construir la funcion para traducir la secuencia.
4. Imprimir el resultado.

```python
#leer archivo
leer_archivo = open("Genoma.txt", "r")
secuencia = leer_archivo.readlines()
leer_archivo.close()
print(secuencia)
```

```python
#Para quitar saltos de lineas (procesar archivo)
secuencia_procesada = ""
for i in secuencia:
  secuencia_procesada += i.replace('\n','')
print(secuencia_procesada)
```

```python
#Primer paso. Crear diccionario

codigo_genetico = {
    'TTT': 'F', 'TTC': 'F', 'TTA': 'L', 'TTG': 'L', 'CTT': 'L', 'CTC': 'L', 'CTA': 'L', 'CTG': 'L',
    'ATT': 'I', 'ATC': 'I', 'ATA': 'I', 'ATG': 'M', 'GTT': 'V', 'GTC': 'V', 'GTA': 'V', 'GTG': 'V',
    'TCT': 'S', 'TCC': 'S', 'TCA': 'S', 'TCG': 'S', 'CCT': 'P', 'CCC': 'P', 'CCA': 'P', 'CCG': 'P',
    'ACT': 'T', 'ACC': 'T', 'ACA': 'T', 'ACG': 'T', 'GCT': 'A', 'GCC': 'A', 'GCA': 'A', 'GCG': 'A',
    'TAT': 'Y', 'TAC': 'Y', 'TAA': '*', 'TAG': '*', 'CAT': 'H', 'CAC': 'H', 'CAA': 'Q', 'CAG': 'Q',
    'AAT': 'N', 'AAC': 'N', 'AAA': 'K', 'AAG': 'K', 'GAT': 'D', 'GAC': 'D', 'GAA': 'E', 'GAG': 'E',
    'TGT': 'C', 'TGC': 'C', 'TGA': '*', 'TGG': 'W', 'CGT': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R',
    'AGT': 'S', 'AGC': 'S', 'AGA': 'R', 'AGG': 'R', 'GGT': 'G', 'GGC': 'G', 'GGA': 'G', 'GGG': 'G',

}

#para comprobar que me arroja la letra del codon
print(codigo_genetico['AAC'])
```

```python
traduccion = ""
for i in range (1,len(secuencia_procesada),3):
  codon = secuencia_procesada[i:i+3]
  if codon in codigo_genetico:
    aminoacidos = codigo_genetico[codon]
    traduccion += aminoacidos
print(traduccion)
```
